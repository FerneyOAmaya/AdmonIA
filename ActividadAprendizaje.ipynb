###1.3. Actividades que el estudiante desarrollará en el curso  (*Diligenciar*)
Incluir una lista de actividades que el estudiante desarrollará dentro del curso, indicando para cada actividad:
- Tiempo en que el estudiante realizará la actividad
- Actividad que realizará el estudiante
- Recurso que desarrollará el docente
- Producto que entregará el estudiante

###1.4. Recurso desarrollado por el docente y herramientas  (*Diligenciar*)
Indicar cómo desarrolará los recursos de la actividad propuesta y que herramienta empleará para hacerlo.

###1.5. Hacerlo atractivo (*Diligenciar*)

Presente un escenario atractivo: Presente un problema o desafío del mundo real que requiera que los estudiantes apliquen sus conocimientos.



###1.6. Métricas y evidencias (*Diligenciar*)

Defina qué aspectos quiere mejorar con la actividad de aprendizaje.

A partir de esos aspectos a mejorar, establezca métricas para medir el éxito de la actividad.

Defina cómo recolectará las evidencias que le permitan evaluar las métricas.

###1.7. Entregables y momentos evaluación y realimentación (*Diligenciar*)

Defina claramente los entregables con sus criterios de evaluación.

Defina momentos de evaluación, defina si es formativa o sumativa.

Defina momentos de realimentación.
